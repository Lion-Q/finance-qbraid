# Set parameters and stock data-random

```python

# parameters for assets and risk factor
num_assets = 4     # number of assets
q = 0.5            # risk factor (risk tolerance of the user)
budget = 3         # set budget
seed = 132     #set random seed

# Generate time series data
stocks = [("STOCK%s" % i) for i in range(num_assets)]
data = RandomDataProvider(tickers=stocks,
                 start=datetime.datetime(1955,11,5),
                 end=datetime.datetime(1985,10,26),
                 seed=seed)
data.run()
```

```python
# plot random finanical data
for (cnt, s) in enumerate(data._tickers):
    plt.plot(data._data[cnt], label=s)
plt.legend()
plt.xticks(rotation=90)
plt.xlabel('days')
plt.ylabel('stock value')
plt.show()
```

```python
# vector with each asset's expected return's mean value per day
mu = data.get_period_return_mean_vector()
```

```python
# covariance matrix Σ（sigma）
sigma = data.get_period_return_covariance_matrix() #Returns the covariance matrix of the four assets
print(sigma)
fig, ax = plt.subplots(1,1)
im = plt.imshow(sigma, extent=[-1,1,-1,1])
x_label_list = ['stock3', 'stock2', 'stock1', 'stock0']
y_label_list = ['stock3', 'stock2', 'stock1', 'stock0']
ax.set_xticks([-0.75,-0.25,0.25,0.75])
ax.set_yticks([0.75,0.25,-0.25,-0.75])
ax.set_xticklabels(x_label_list)
ax.set_yticklabels(y_label_list)
plt.colorbar()
plt.clim(-0.000002, 0.00001)
plt.show()
```

#Create the portfolio instance using PortfolioOptimization class

```python
from qiskit_finance.applications.optimization import PortfolioOptimization
from qiskit_optimization.algorithms import MinimumEigenOptimizer
```

```python
portfolio = portfolio = PortfolioOptimization(expected_returns=mu, covariances=sigma, risk_factor = q, budget=budget)
qp = portfolio.to_quadratic_program()
```

```python
#classical optimizer
exact_mes = NumPyMinimumEigensolver()
exact_eigensolver = MinimumEigenOptimizer(exact_mes)
result = exact_eigensolver.solve(qp)

print(result)  # if x_n = 1.0 : buy, if x_n = 0.0 : do not buy
```

#Fetch stocks by market cap for both S&P 500 and Nasdaq

```python
import yfinance as yf
import pandas as pd


# Define the tickers of stocks
sp500_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'NVDA', 'BRK-B', 'META', 'V', 'JNJ']
nasdaq_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'TSLA', 'META', 'PYPL', 'ADBE', 'INTC']

# Combine tickers
tickers = sp500_tickers + nasdaq_tickers
```

```python
# Download historical data from 2000 to now
data = yf.download(tickers, start="2000-01-01", end="2024-09-23")['Adj Close']

# Display the first few rows of the data to confirm
print(data.head())
```

```python
# Calculate expected returns and covariance matrix needed for the portfolio optimization
returns = data.pct_change().dropna()
expected_returns = returns.mean()
cov_matrix = returns.cov()
```

```python
# daily values (0.05 ==5%)
print("Expected Returns:\n", expected_returns)
print("Covariance Matrix:\n", cov_matrix)
```

```python
matrix = np.array(cov_matrix)  # convert to a NumPy array
diagonal_elements = np.diag(matrix)

# Find the minimum value in the diagonal
min_diagonal_value = np.min(diagonal_elements)
min_value = np.min(matrix)
```

```python
# Plotting the covariance matrix as a heatmap
fig, ax = plt.subplots(figsize=(10, 8))

# Convert the covariance matrix to a numpy array for plotting
sigma = cov_matrix.values

# Create the heatmap
im = ax.imshow(sigma, cmap='viridis')

# Set custom tick labels based on the stock tickers
x_label_list = cov_matrix.columns.tolist()  # Extract column names for labels
y_label_list = cov_matrix.index.tolist()    # Extract index names for labels
ax.set_xticks(np.arange(len(x_label_list)))
ax.set_yticks(np.arange(len(y_label_list)))
ax.set_xticklabels(x_label_list, rotation=90)
ax.set_yticklabels(y_label_list)


# Calculate min and min_diagonal_value
min_value = np.min(sigma)
min_diagonal_value = np.min(np.diag(sigma))

# Set the color limits on the mappable (heatmap)
im.set_clim(min_value, min_diagonal_value)

# Add color bar with limits set based on matrix values
plt.colorbar(im)

plt.title('Covariance Matrix (Sigma) of Stock Returns')
plt.show()
```

```python
portfolio = portfolio = PortfolioOptimization(expected_returns=expected_returns, covariances=sigma, risk_factor = q, budget=budget)
qp = portfolio.to_quadratic_program()
```

```python
#classical optimizer
exact_mes = NumPyMinimumEigensolver()
exact_eigensolver = MinimumEigenOptimizer(exact_mes)
result = exact_eigensolver.solve(qp)

print(result)
```
